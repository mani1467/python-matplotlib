### Matplotlib Plotting

##### Ploting x and y points
The plot() function is used to draw points (markers) in a diagram.
By default,the plot() function draws a line from point to point.
The function takes parameters for specifying points in the diagram.
parameter 1 is an array containing the points on the x-axis.
parameter 2 is an array containing the point on the y-axis.

if we need to plot a line from(1,3)to(8,10) we have to pass two arrays[1,8]and [3,10]to the plot function.

```python
#draw a line in a diagram from position (1,3) to position (8,10)
import matplotlib.pyplot as plt
import numpy as np
xpoints=np.array([1,8])
ypoints=np.array([3,10])
plt.plot(xpoints,ypoints)
plt.show()
```

#### Matplotlib labels and Title:

#### creating a Title for a Plot:
with pyplot you can use the title() function to set a title for the plot.

```python
# a plot title and labels for the x- and y-axis:
x=np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y=np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
plt.plot(x,y)
plt.xlabel("average pulse")
plt.ylabel("calorie Burnage")
plt.title("sport watch data")
plt.show()
```

#### Set Font Properties for title and labels:

use the "fontdict" parameter in xlabel(),ylabel() and title() to set font properties for the title and labels.

```python
#Set font properties for the title and labels:

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1={"family":"serif","color":"blue","size":20}
font2={"family":"serif","color":"darkred","size":15}
plt.title("sport watch data",fontdict=font1)
plt.xlabel("average pluse",fontdict=font2)
plt.ylabel("calories burnage",fontdict=font2)
plt.plot(x,y)
plt.show()
```

#### Position the title:
Use the 'loc'parameter in "title" to the position the title.
Legal values are:'left','right' and 'center'.defult value is center.

```python
#Position the title to the right:

x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])
font1={"family":"serif","color":"blue","size":20}
font2={"family":"serif","color":"darkred","size":15}
plt.title("sport watch data",fontdict=font1,loc='right')
plt.xlabel("average pluse",fontdict=font2)
plt.ylabel("calories burnage",fontdict=font2)
plt.plot(x,y)
plt.show()
```

####  Matplotlib Scatter:
###### Creating scatter plots
you can use the scatter() function to draw a scatter plot.

The scatter() function plots one dots for each observation.it needs two arrays of the same length,one for the values of the x-axis and one for values on the y-axis.

```python
#a simple scatter plot
import matplotlib.pyplot as plt
import numpy as np
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x,y)
plt.show()

#the x-axis shows how old the car is.
#the y-axis shows the speed of the car when it passes.
```

##### Compare plots:
In the example above, there seems to be a relationship between speed and age, but what if we plot the observations from another day as well? Will the scatter plot tell us something else?

```python
#draw two plots on the same figure:
#day-1, the age and speed of 13 cars:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

#day-2,the age and speed of 15 cars;
x=np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y=np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x,y)

plt.show()
```

###### by comparing the two plots.the two plots are plotted with two differentcolors then easly visual to user.set own colors are apply.

```python
#You can set your own color for each scatter plot with the color or the c argument
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y, color = 'hotpink')

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y, color = '#88c999')

plt.show()
```

##### Color each dot

##### ColorMap
The matplotlib module has a number of available colormaps.
A colormap is like a list of color, where each color has a value that range from 0 to 100.
##### how to use the colormap
You can specify the colormap with the keyword argument "cmap" with the value of the colormap, in this case 'viridis' which is one of the built-in colormaps available in Matplotlib.

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x,y,c=colors,cmap='viridis')

plt.show()
```

###### you can include the colormap in the drawing by including the "plt.colorbar()" statement:

```python
#include the actual colormap:
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0,10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
plt.scatter(x,y,c=colors,cmap='viridis')
plt.colorbar()
plt.show()
```

##### the hist() function will erad the array and produce a histogram:

```python
#simple histogram:
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)

plt.hist(x)
plt.show() 
```

### Matplotlib Pie charts

##### creating pie charts
Use the pie() function to draw pie charts.

```python
# simple pie chart
y=np.array([20,35,30,25,20,15])
plt.pie(y)
plt.show()
```

#### Labels
Add labels to the pie chart with the labels parameter.

The labels parameter must be an array with one label for each wedge:


```python
y=np.array([40,35,30,25,20,15])
mylabels=['apple','banana','orange','grapes','dates','mango']
plt.pie(y,labels=mylabels)
plt.show()
```

### Start Angle:
As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a startangle parameter.

The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
#start the first wedge at 180 degrees:
y=np.array([40,35,30,25,])
mylabels=['apple','banana','orange','grapes']
plt.pie(y,labels=mylabels,startangle=180)
plt.show()
```

## Explode
Maybe you want one of the wedges to stand out? The explode parameter allows you to do that.

The explode parameter, if specified, and not None, must be an array with one value for each wedge.

Each value represents how far from the center each wedge is displayed:

```python
# pull the banana wedge 0.2 from the center of the pie.
y=np.array([40,35,30,25,])
mylabels=['apple','banana','orange','grapes']
myexplode=[0,0.2,0,0]
plt.pie(y,labels=mylabels,startangle=180 , explode=myexplode)
plt.show()
```

## Shadow
add a shadow to the pie chart by setting the "shadow()" parameter to true.

```python
y=np.array([40,35,30,25,])
mylabels=['apple','banana','orange','grapes']
myexplode=[0,0.1,0.1,0]
plt.pie(y,labels=mylabels,startangle=180 , explode=myexplode,shadow=True)
plt.show()
```

## Colors:
You can set the color of each wedge with the colors parameter.

The colors parameter, if specified, must be an array with one value for each wed.e:

```python
#specify a new color of each wedge.
#so many color we can use like hexadecimal colors and supporting colors you can use it.
y=np.array([40,35,30,25,])
mylabels=['apple','banana','orange','grapes']
mycolors=['black','blue','red','green']
plt.pie(y,labels=mylabels,colors=mycolors)
plt.show()
```

## Legend
to add a list of explanation for each wedge,use the legend() function.

```python
#add a legend
y=np.array([40,35,30,25,])
mylabels=['apple','banana','orange','grapes']
plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

### Legend With header
header to the legend like title

```python
y=np.array([40,35,30,25,])
mylabels=['apple','banana','orange','grapes']
plt.pie(y,labels=mylabels)
plt.legend(title='four fruits:')
plt.show()
```
